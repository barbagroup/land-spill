```python
import numpy
from matplotlib import pyplot
```

```python
Re = numpy.logspace(1, 6, 1000, dtype=numpy.float64)
```

```python
'''
Three-regime friction coefficient calculation for open-channel
'''

three_regimes_open_channel = numpy.zeros_like(Re)

temp = (Re <= 500)
three_regimes_open_channel[temp] = 24. / Re[temp]

temp = numpy.logical_and(Re>500, Re<=1250)
three_regimes_open_channel[temp] = 0.224 / (Re[temp]**0.25)

temp = (Re > 1250)
three_regimes_open_channel[temp] = 0.25 / (numpy.log10(1.64833821394207454/(Re[temp]**0.9))**2)
```

```python
'''
Three-regime friction coefficient calculation for full-flow circular pipe
'''

three_regimes_circular_pipe = numpy.zeros_like(Re)

temp = (Re <= 500)
three_regimes_circular_pipe[temp] = 16. / Re[temp]

temp = numpy.logical_and(Re>500, Re<=1250)
three_regimes_circular_pipe[temp] = 0.224 / (Re[temp]**0.25)

temp = (Re > 1250)
three_regimes_circular_pipe[temp] = 0.25 / (numpy.log10(1.64833821394207454/(Re[temp]**0.9))**2)
```

```python
'''
Churchill's friction coefficient model
'''
churchill = numpy.zeros_like(Re)
theta1 = (2.457 * numpy.log((1.75/Re)**(-0.9)))**16
theta2 = (37530./4./Re)**16
churchill = 8. * (((2./Re)**12  + 1./((theta1 + theta2)**1.5))**(1./12))
```

```python
pyplot.figure(figsize=(8, 6), dpi=300)
pyplot.loglog(Re, three_regimes_open_channel,
              linestyle='-', lw=3,
              label="Three-regime calculation (wide open channel)")
pyplot.loglog(Re, three_regimes_circular_pipe,
              linestyle='-.', lw=2,
              label="Three-regime calculation (full-flow pipe)")
pyplot.loglog(Re, churchill,
              linestyle='--', lw=1,
              label="Churchill's model")
pyplot.ylim(10**(-2.5), 1.)
pyplot.legend(loc=0)
pyplot.xlabel(r"$Re$" + " (defined by hydraulic radius)")
pyplot.ylabel("Darcy-Weisbach coefficient")
pyplot.title("Darcy-Weisbach friction coefficient, "+r"$\epsilon/D=0$")
pyplot.grid()
```

```python
'''
Three-regime friction coefficient calculation for full-flow circular pipe, eps/D=0.02
'''

three_regimes_circular_pipe = numpy.zeros_like(Re)

temp = (Re <= 500)
three_regimes_circular_pipe[temp] = 16. / Re[temp]

temp = numpy.logical_and(Re>500, Re<=1250)
three_regimes_circular_pipe[temp] = 0.224 / (Re[temp]**0.25)

temp = (Re > 1250)
three_regimes_circular_pipe[temp] = 0.25 / (numpy.log10(0.02/3.7+1.64833821394207454/(Re[temp]**0.9))**2)
```

```python
'''
Churchill's friction coefficient model, eps/D=0.02
'''
churchill = numpy.zeros_like(Re)
theta1 = (2.457 * numpy.log(1./((1.75/Re)**0.9+0.27*0.02)))**16
theta2 = (37530./4./Re)**16
churchill = 8. * (((2./Re)**12  + 1./((theta1 + theta2)**1.5))**(1./12))
```

```python
pyplot.figure(figsize=(8, 6), dpi=300)
pyplot.loglog(Re, three_regimes_open_channel,
              linestyle='-', lw=3,
              label="Three-regime calculation (wide open channel)")
pyplot.loglog(Re, three_regimes_circular_pipe,
              linestyle='-.', lw=2,
              label="Three-regime calculation (full-flow pipe)")
pyplot.loglog(Re, churchill,
              linestyle='--', lw=1,
              label="Churchill's model")
pyplot.ylim(10**(-2.5), 1.)
pyplot.legend(loc=0)
pyplot.xlabel(r"$Re$" + " (defined by hydraulic radius)")
pyplot.ylabel("Darcy-Weisbach coefficient")
pyplot.title("Darcy-Weisbach friction coefficient, "+r"$\epsilon/D=0.02$")
pyplot.grid()
```
